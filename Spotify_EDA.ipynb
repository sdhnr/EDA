## Data Analysis

### Top 5 Most Popular Artists

```python
top_five_artists=df.groupby('artist').count().sort_values(by='song_title',ascending=False)['song_title'][:5]
top_five_artists
```

```python
top_five_artist=df["artist"].groupby(df["artist"]).count().nlargest(5)
top_five_artist
```

```python
top_five_artists.plot.barh()
plt.grid(axis='x',linestyle='--');
```

### Top 5 Loudest Tracks

```python
top_five_loudest = df[['loudness','song_title']].sort_values(by='loudness',ascending=True)[:5]
```

```python
top_five_loudest
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='loudness',y='song_title',data=top_five_loudest)
plt.title('Top 5 Loudest Tracks')
plt.grid(axis='x',linestyle='--')
plt.xlabel('Loudness')
plt.ylabel('Song Title');
```

### Artist with the most danceability song

```python
top_five_artist_danceability = df[['danceability','song_title','artist']].sort_values(by='danceability',ascending=False)[:5]
top_five_artist_danceability
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='danceability',y='artist',data=top_five_artist_danceability);
plt.title('Artist With The Most Danceability Song')
plt.xlabel('Danceability')
plt.ylabel('Artist')
plt.grid(axis='x',linestyle='--');
```

### Top 10 Instrumentalness tracks

```python
df.head(5)
```

```python
top_ten_intrumental = df[['instrumentalness','song_title','artist']].sort_values(by='instrumentalness',ascending=False)[:10]
top_ten_intrumental
```

```python
plt.figure(figsize=(10,5))
plt.pie(x='instrumentalness',data=top_ten_intrumental,labels=top_ten_intrumental.song_title,autopct='%1.2f')
plt.title('Top Ten Instrumental Tracks');
```

### Multiple Feature Plots

```python
df.columns
```

```python
interest_feature_columns = ['tempo','loudness','acousticness','danceability','duration_ms','energy','instrumentalness',
                            'liveness','speechiness','valence']
```

```python
for feature_col in interest_feature_columns:
    pos_data = df[df['target']==1][feature_col]
    neg_data = df[df['target']==0][feature_col]
    
    plt.figure(figsize=(10,5))
    
    sns.distplot(pos_data,bins=30,label='Positive',color='g')
    sns.distplot(neg_data,bins=30, label='Negative', color='r')
    
    plt.legend(loc='upper right')
    plt.title(f'Positive and Negative Histogram Plot for {feature_col}');
```
