# 2 Team Names

```python
team = list(pd.read_html('https://en.wikipedia.org/wiki/National_Basketball_Association')[3]['Team'].values.flatten())
```

```python
team.remove('Western Conference') #In my team list there is one item other than team names ,so I will remove it.
```

```python
len(team)  # There are 30 teams in NBA
```

# 3 Team Roster dict

```python
dict_num_player = {}
dict_team_roster = {}
for i in range(len(team)):
    dict_team_roster[team[i]] = nba[2*i+1]
    dict_team_roster[team[i]]['Team'] = team[i]
    dict_num_player[team[i]] = len(nba[2*i+1])
    dict_team_roster[team[i]].rename({'Pos.':'Pos','No.':'No','DOB (YYYY-MM-DD)':'DOB'},axis=1,inplace=True)
```

```python
dict_num_player  # 
```

# 4 Concat rosters

```python
df = pd.DataFrame(columns=dict_team_roster[team[0]].columns)
for i in range(len(team)):
    df = pd.concat([df,dict_team_roster[team[i]]],axis=0)
df
```

```python
df.reset_index(drop = True ,inplace=True)
df
```

# 7 Height Column

```python
df.loc[0,'Height']  # Trying to find the type of values in Height column
```

```python
print(df.loc[0,'Height'])
```

```python
def meter(s):
    ind = s.find('(')                     # Define a function to convert height values to a float 
    return float(s[ind+1:-3])
```

```python
meter('6\xa0ft 4\xa0in (1.93\xa0m)')
```

```python
meter(df.loc[3,'Height'])
```

```python
df['Height'] = df['Height'].apply(meter)

```

```python
df.head()
```

# 8 Weight

```python
df.loc[0,'Weight']
```

```python
def kg(s):
    ind = s.find('(')
    return int(s[ind+1:-4])        #Convert weight to kilogram(kg) 
```

```python
df['Weight'] = df['Weight'].apply(kg)
```
